<h1>Data Preparation</h1>

<h3>Importing needed Libraries</h3>

```python
import os
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

<h3>Importing all the Data Files</h3>

```python
data_folder = '.\\CW_Data'

data_list = os.listdir(data_folder)
data_list
```

<h3> Combining all the Data Files into a DataFrame </h3>

```python
combined_data = []

for each in data_list:
    if each.endswith(".csv"):
        df = pd.read_csv(os.path.join(data_folder, each))
        combined_data.append(df)
        
combined_df = pd.concat(combined_data)
combined_df
```

<h3> Removing all the NaN values </h3>

```python
combined_df = combined_df.dropna()
combined_df
```

<h3> Converting Datatypes of Columns </h3>

```python
combined_df[['Quantity Ordered','Price Each']].apply(pd.to_numeric)
combined_df.dtypes
```

<h3> Creating new Column "Month" from "Order Date" </h3>

```python
combined_df['Order Date'] = pd.to_datetime(combined_df['Order Date'])
combined_df['Month'] = combined_df['Order Date'].dt.month
combined_df
```

<h3>Creating new Column "City" from "Purchase Address"</h3>

```python
combined_df['City']  = [i.split(",")[1] for i in combined_df['Purchase Address'].values]
combined_df
```

<h1>Data Analysis</h1>


<h3> Sum of total sales made in each Month </h3>


```python
combined_df['Total Price'] = combined_df['Price Each']*combined_df['Quantity Ordered']
Monthly = combined_df[['Month','Total Price']]
Monthly_sales = Monthly.groupby('Month')['Total Price'].sum()
Monthly_sales_df = pd.DataFrame(Monthly_sales)
Monthly_sales_df
```

<h3> Mean of Total Sales made in Each City </h3>

```python
City_orders = combined_df[['City','Total Price']]
City_orders = City_orders.groupby('City')['Total Price'].mean()
Total_city_Orders = pd.DataFrame(City_orders)
Total_city_Orders
```

<h3> Standard Deviation </h3>

```python
SD = combined_df[['Quantity Ordered','Total Price']].std()
SD = pd.DataFrame(SD)
SD
```

<h3> Skewness </h3>

```python
SK = combined_df[['Quantity Ordered','Price Each']].skew()
SK = pd.DataFrame(SK)
SK
```

<h3> Kurtosis </h3>

```python
KS = combined_df[['Quantity Ordered','Total Price']].kurtosis()
KS = pd.DataFrame(KS)
KS
```

<h3> Finding Correlation Between Columns </h3>

```python
matrix = combined_df.corr()
sns.heatmap(matrix)
plt.show()
```

<h1>Data Exploration</h1>

<h3>Month that made Most Sales</h3>

```python
Monthly = combined_df[['Month','Total Price']]
Monthly_sales = Monthly.groupby('Month')['Total Price'].sum()
Monthly_sales_df = pd.DataFrame(Monthly_sales)
Monthly_sales_df
```

```python
Monthly_sales.plot.bar(x='Month', y='Total Price', title='Total Sales of Each Month')
plt.show()
```

<h3> City that made the most Sales </h3>

```python
City_orders = combined_df[['City','Quantity Ordered']]
City_orders = City_orders.groupby('City')['Quantity Ordered'].sum()
Total_city_Orders = pd.DataFrame(City_orders)
Total_city_Orders
```

```python
Total_city_Orders.idxmax().max()
```

<h3> Most sold Products </h3>

```python
Products = combined_df[['Product','Quantity Ordered']]
Total_Products = Products.groupby('Product')['Quantity Ordered'].sum()
Total_Products_df = pd.DataFrame(Total_Products)
Total_Products_df
```

```python
Total_Products_df.idxmax().max()
```

```python
Total_Products.plot.bar(x='Product',y='Quantity Ordered',title='Total sales of each item ordered')
plt.show()
```
